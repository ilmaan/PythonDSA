```python
def createlinkedlist(li):
    head = None
    current = None

    for i in li:
        if i == -1:
            break

        # Create a new node as a dictionary
        new_node = {'value': i, 'next': None}

        if head is None:
            # If the list is empty, set head to the new node
            head = new_node
            current = new_node
        else:
            # Otherwise, link the current node to the new node
            current['next'] = new_node
            current = new_node

    return head

# Helper function to print the linked list
def print_linkedlist(head):
    current = head
    while current is not None:
        print(current['value'], end=" -> ")
        current = current['next']
    print("None")

# Example usage
li = [1, 2, 3, 4, -1, 5]
head = createlinkedlist(li)
print_linkedlist(head)
```

```python
class linklist:
    def __init__(self, val):
        self.data = val
        self.next = None

    def zipll(self, v1, v2):
        # Initialize pointers
        curr1 = v1
        curr2 = v2
        dummy = linklist(0)
        curr = dummy

        # Flag to alternate between lists
        flag = 1

        # Loop until one of the lists is exhausted
        while curr1 is not None and curr2 is not None:
            print('------>>>',curr.data)
            if flag == 1:
                curr.next = curr1
                curr1 = curr1.next
                flag = 0
                
            else:
                curr.next = curr2
                curr2 = curr2.next
                flag = 1
                
            curr = curr.next

        # Append the remaining elements, if any
        if curr1 is not None:
            curr.next = curr1
        if curr2 is not None:
            curr.next = curr2

        # Return the head of the new list
        return dummy.next

    def print_list(self):
        curr = self
        while curr is not None:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")


a1 = linklist('a')
b1 = linklist('b')
c1 = linklist('c')
d1 = linklist('d')

a1.next = b1
b1.next = c1
c1.next = d1

a2 = linklist('1')
b2 = linklist('2')
c2 = linklist('3')
d2 = linklist('4')
e2 = linklist('5')
f2 = linklist('6')

a2.next = b2
b2.next = c2
c2.next = d2
d2.next = e2
e2.next = f2

# Zipping the two lists
zipped_list = a1.zipll(a1, a2)

# Print the resulting list
zipped_list.print_list()
```

Remove Duplicate from a linked list

```python
class linklist:
    def __init__(self, val):
        self.data = val
        self.next = None

    def removeduplicate(self, head):
        if head is None:
            return False
        while head.next is not None:
            if head.data == head.next.data:
                head.next = head.next.next
            else:
                head = head.next
        
    
    def print_list(self, head):
        curr = head
        while curr is not None:
            print(curr.data)
            curr = curr.next
        print("NONE")
    


# Create linked list nodes
a1 = linklist('a')
b1 = linklist('a')
c1 = linklist('b')
d1 = linklist('b')

# Link nodes
a1.next = b1
b1.next = c1
c1.next = d1

print("Original list:")
# a1.print_list(a1)

a1.removeduplicate(a1)

print("Removed duplicate:")
a1.print_list(a1)
```

Palindrome Linked List

```python
class linklist:
    def __init__(self, val):
        self.data = val
        self.next = None

    def find_middle(self,head):
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        print("SLOW",slow.data)
        return slow
    
    def reverse(self,head):
        prev = None
        current = head
        while current:
            next = current.next
            current.next = prev
            prev = current
            current = next
             
        return prev
    
    def is_palindrome(self,head):

        if head is None or head.next is None:
            return True
        
        middle = self.find_middle(head)

        second_hald = self.reverse(middle)

        first_half = head

        while first_half and second_hald:
            if first_half.data != second_hald.data:
                return False
            first_half = first_half.next
            second_hald = second_hald.next
        return True
    
    def print_list(self,head):
        current = head
        while current:
            print(current.data)
            current = current.next
            
# Create linked list nodes
a1 = linklist(1)
b1 = linklist(2)
c1 = linklist(1)
d1 = linklist(2)
e1 = linklist(3)
# f1 = linklist(3)

# Link nodes
a1.next = b1
b1.next = c1
c1.next = d1
d1.next = e1
# e1.next = f1

print("Original list:")
a1.print_list(a1)

# a1.linklist(a1)
# a1.is_palindrome(a1)

if a1.is_palindrome(a1):
    print("Linked list is palindrome")
    
else:
    print("Linked list is not palindrome")

print("Linked list:")


# print("Removed duplicate:")
a1.print_list(a1)
```
