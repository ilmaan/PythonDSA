```python
def bubsort(li):
    for i in range(len(li)-1):
        # print('----',len(li)-1)
        for j in range(len(li)-1-i):
            # print(len(li)-1-i)
            # print(li[j],li[j+1])
            if (li[j]>li[j+1]):
                li[j],li[j+1]=li[j+1],li[j]
                print(li)
```
